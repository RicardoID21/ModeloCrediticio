```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("train-3.csv")
df.info()
dfraw = df
```

```python
# Aplica .strip() a todas las columnas tipo texto y convierte "" a NaN
obj_cols = df.select_dtypes(include=["object"]).columns.tolist()

for c in obj_cols:
    # Solo si la serie no es todo NaN
    s = df[c].astype("object")
    s = s.where(~s.isna(), None)
    s = s.apply(lambda x: x.strip() if isinstance(x, str) else x)
    # VacÃ­os a NaN
    s = s.replace("", np.nan)
    df[c] = s

df.head()
```

```python
drop_id_cols = ["ID", "Name", "SSN"]
target_col   = "Credit_Score"
time_col     = "Month" 

# X: solo variables candidatas a modelar
train = df.drop(columns=drop_id_cols + [target_col, time_col], errors="ignore").copy()

# y: etiqueta
credit = df[target_col].copy()

# t: Ã­ndice temporal para particionar train/valid por fecha/mes
t = df[time_col].copy()

train.head()
```

```python
def mode_float(s):
    s = s.dropna().astype(float)
    if s.empty:
        return np.nan
    vc = s.value_counts()
    top = vc[vc == vc.max()].index.astype(float).tolist()
    return min(top)  # si hay empate, toma el menor

def mode_int(s):
    s = s.dropna()
    if s.empty:
        return np.nan  # mantiene NaN si no hay valores
    vc = s.value_counts()
    top = vc[vc == vc.max()].index.tolist()
    return int(min(top)) 

def mode_cat(s: pd.Series):
    s = s.dropna()
    return s.mode().iat[0] if not s.mode().empty else np.nan
```

```python
#Age

train["Age"] = (
    train["Age"].astype(str)          # por si viene como object/mixto
            .str.replace("_", "", regex=False)
            .str.strip()
)
train["Age"] = pd.to_numeric(train["Age"], errors="coerce")
train.loc[train["Age"] < 0, "Age"] = np.nan
train.loc[train["Age"] > 120, "Age"] = np.nan
train["Age"] = pd.to_numeric(train["Age"], errors="coerce")
print("n NaN en Age:", train["Age"].isna().sum())
    
# Calculamos el modo de Age por Customer_ID
age_mode = train.groupby("Customer_ID")["Age"].transform(mode_float)

train["Age"] = np.where(train["Age"].isna(), age_mode, train["Age"])
train["Age"] = np.where(age_mode.notna(), age_mode, train["Age"])

print(train["Age"].describe())
```

```python
train["Occupation"] = (
    train["Occupation"].astype(str)
    .str.lower()
    .str.replace("_", " ", regex=False)   # 'media_manager' -> 'media manager'
    .str.strip()
    .replace({"": np.nan, "_______": np.nan, "na": np.nan, "n/a": np.nan, "none": np.nan})
)

occ_mode = train.groupby("Customer_ID")["Occupation"].transform(mode_cat)
train["Occupation"] = train["Occupation"].fillna(occ_mode)

bucket_map = {
    # self-employed / emprendedor
    "entrepreneur": "self_employed",
    "self employed": "self_employed",
    "businessman": "self_employed",

    # STEM / tÃ©cnicos profesionales
    "developer": "stem_professional",
    "engineer": "stem_professional",
    "scientist": "stem_professional",
    "architect": "stem_professional",

    # salud
    "doctor": "health_professional",
    "nurse": "health_professional",   # por si aparece

    # white-collar legal/finanzas/gestiÃ³n
    "lawyer": "white_collar",
    "accountant": "white_collar",
    "manager": "white_collar",

    # oficios
    "mechanic": "skilled_trade",

    # educaciÃ³n / medios / artes
    "teacher": "edu_media_arts",
    "journalist": "edu_media_arts",
    "writer": "edu_media_arts",
    "musician": "edu_media_arts",
    "media manager": "edu_media_arts",
}

mapped = train["Occupation"].map(bucket_map)

# 3) Sobrescribe Occupation con UNA sola columna:
#    - Si estaba nulo -> 'unknown'
#    - Si no estÃ¡ en el mapa -> 'other'
train["Occupation"] = mapped.where(train["Occupation"].notna(), "unknown").fillna("other")

# Chequeo rÃ¡pido
print(train["Occupation"].value_counts(dropna=False).head(10))

train["Occupation"].unique()
train.head()
```

```python
train.loc[train["Num_Bank_Accounts"] < 0, "Num_Bank_Accounts"] = np.nan
train.loc[train["Num_Bank_Accounts"] > 11, "Num_Bank_Accounts"] = np.nan
sal_mode = train.groupby("Customer_ID")["Num_Bank_Accounts"].transform(mode_int)
train["Num_Bank_Accounts"] = train["Num_Bank_Accounts"].fillna(sal_mode)
train["Num_Bank_Accounts"].value_counts()
```

```python
#Monthly Inhand Salary
#En este paso rellenamos los valores faltantes de Monthly_Inhand_Salary usando el valor mÃ¡s frecuente (modo) de esa misma 
# variable dentro de cada Customer_ID, de modo que cada cliente conserve un salario mensual consistente en sus filas.

sal_mode = train.groupby("Customer_ID")["Monthly_Inhand_Salary"].transform(mode_float)
train["Monthly_Inhand_Salary"] = train["Monthly_Inhand_Salary"].fillna(sal_mode)

train.head()
```

```python
#Num_Credit_Card
train["Num_Credit_Card"].value_counts()
train.loc[train["Num_Credit_Card"] < 0, "Num_Credit_Card"] = np.nan
train.loc[train["Num_Credit_Card"] > 11, "Num_Credit_Card"] = np.nan

sal_mode = train.groupby("Customer_ID")["Num_Credit_Card"].transform(mode_int)
train["Num_Credit_Card"] = train["Num_Credit_Card"].fillna(sal_mode)

train.head()
```

```python
#Interest_Rate convertimos a porcentaje
train.loc[train["Interest_Rate"] < 0, "Interest_Rate"] = np.nan
train.loc[train["Interest_Rate"] > 34, "Interest_Rate"] = np.nan

sal_mode = train.groupby("Customer_ID")["Interest_Rate"].transform(mode_int)
train["Interest_Rate"] = train["Interest_Rate"].fillna(sal_mode)
train["Interest_Rate"] = train["Interest_Rate"] / 100

train.head()
```
